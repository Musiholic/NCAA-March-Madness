```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# read in files
super_data_df = pd.read_csv("Resources/madness_label.csv")

super_data_df
```

```python
#seperate the x and y variables

y = super_data_df["Madness"]
X = super_data_df.drop(columns="Madness")
```

```python
X = X.drop(columns="Team")
```

```python
display(y)
```

```python
display(X)
```

```python
X.PTS = X.PTS.apply(lambda x : x.replace(',',''))
X.DRebs = X.DRebs.apply(lambda x : x.replace(',',''))
X.REB = X.REB.apply(lambda x : x.replace(',',''))
X.FGM_y = X.FGM_y.apply(lambda x : x.replace(',',''))
X.FGA_y = X.FGA_y.apply(lambda x : x.replace(',',''))
X.FTA = X.FTA.apply(lambda x : x.replace(',',''))
```

```python
X
```

```python
df = X[X['Years'] == 2324].sort_values(by=['PTS'], ascending=False)
```

```python
X = X.drop(columns="Years")
```

```python
X['PTS'] = X['PTS'].astype(float)
X['DRebs'] = X['DRebs'].astype(float)
X['REB'] = X['REB'].astype(float)
X['FGM_y'] = X['FGM_y'].astype(float)
X['FGA_y'] = X['FGA_y'].astype(float)
X['FTA'] = X['FTA'].astype(float)
```

```python
X.dtypes
```

```python
# Split the data using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
lr_model = LogisticRegression(random_state=1)

# Fit the model using training data
lr_model.fit(X_train, y_train)
```

```python
# Make a prediction using the testing data
lr_prediction = lr_model.predict(X_test)
```

```python
# Generate a confusion matrix for the model
cm_imbalanced = confusion_matrix(y_test, lr_prediction)
cm_imbalanced_df = pd.DataFrame(cm_imbalanced)
cm_imbalanced_df
```

```python
# Print the classification report for the model
print(classification_report(y_test, lr_prediction))
```
