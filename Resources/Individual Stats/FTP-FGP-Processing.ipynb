```python
directoryFTP = "/Users/jaxontige/Documents/Data Analytics Bootcamp/Projects/Project 4/Data/FTP/FTP CSVs"
all_FTP_dfs = []

# Loop through each file in the directory
for filename in os.listdir(directoryFTP):
    if filename.endswith('.csv'):  # Assuming all files are CSV files
        filepath = os.path.join(directoryFTP, filename)
        # Extract the four-digit code from the filename
        code = filename[:5]
        # Read CSV file and add a column for the code
        df = pd.read_csv(filepath)
        df['Years'] = code  # Add a new column for the code
        all_FTP_dfs.append(df)

        # combine all dfs into one 
combined_df_FTP = pd.concat(all_FTP_dfs, ignore_index=True)
combined_df_FTP['Years'] = combined_df_FTP['Years'].str.replace('-','')

# combined_df_atr.to_csv("combined_assist_and_turnover_ratio.csv")
```
