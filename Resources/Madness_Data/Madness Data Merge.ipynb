```python
directoryMADNESS = "/Users/jaxontige/Documents/Data Analytics Bootcamp/Projects/Project 4/Data/Madness Data"
all_Madness_dfs = []

# Loop through each file in the directory
for filename in os.listdir(directoryMADNESS):
    if filename.endswith('.csv'):  # Assuming all files are CSV files
        filepath = os.path.join(directoryMADNESS, filename)
        # Extract the four-digit code from the filename
        code = filename[:5]
        # Read CSV file and add a column for the code
        df = pd.read_csv(filepath)
        df['Years'] = code  # Add a new column for the code
        all_Madness_dfs.append(df)

        # combine all dfs into one 
combined_df_MADNESS = pd.concat(all_Madness_dfs, ignore_index=True)

# combined_df_atr.to_csv("combined_assist_and_turnover_ratio.csv")
```
