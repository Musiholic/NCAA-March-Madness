```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
super_data_cleaned = pd.read_csv("Resources/super_data_cleaned.csv")
combined_madness = pd.read_csv("Madness_Data/Combined_Madness.csv")
```

```python
combined_madness
```

```python
super_data_cleaned
```

```python
safe = super_data_cleaned.Years.unique()
[print(x)for x in safe]
```

```python
empty_data = pd.DataFrame()
```

```python
for year in safe:
    df = super_data_cleaned[super_data_cleaned['Years'] == year].sort_values(by=['PTS'], ascending=False)
    top_150 = df.head(150)
    empty_data = pd.concat([empty_data, top_150])
```

```python
empty_data
```

```python
balanced_data = empty_data
```

```python
merged_data = combined_madness.merge(balanced_data, left_on=['Team','Years'], right_on=['Team','Years'], how='right')
```

```python
merged_data
```

```python
df_filled = merged_data.fillna(0)
```

```python
df_filled
```

```python
df_clean = df_filled.drop({"Unnamed: 0_x", "Rank", "FG%_y", "GM", "W-L", "FGM_x", "FGA_x", "FG%_x", "Unnamed: 0_y"}, axis=1)
df_clean.head(50)
```

```python
df_clean["Madness"].value_counts()
```

```python
# Write combined DataFrame to CSV
df_clean.to_csv('madness_label.csv', index=False)
```
